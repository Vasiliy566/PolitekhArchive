```python
import numpy
from scipy import linalg
import math
import matplotlib.pyplot as plt
%matplotlib inline

p = [1.0, 0.1, 0.01, 0.0001, 0.000001]

A = numpy.matrix([[ 0,  2,  8, -7,  7,  5, -7, -7],
                 [  7,  2, -4,  2,  3,  3, -1, -2],
                 [ -7,  2,  1,  3,  6, -6, -3, -4],
                 [ -2, -8, -6, -1,  6,  2,  1, -4],
                 [  0,  4, -7,  1, 22,  0, -6, -6],
                 [  0, -3, -6,  6,  4, 13,  0,  6],
                 [ -8, -6, -4,  7, -5, -5, -2,  1],
                 [  5,  5, -2, -2, -3,  0, -7, 14]])

b = numpy.matrix([[6.0], [36], [-25], [-57], [32], [62], [-71], [70]])
```
